# Dependencies

```python
import pandas as pd
import os
```

```python
# Read all file from folder './Resources/Raw Data'
file_list = os.listdir('./Resources/NYC-data')
file_list
```

```python
# Read all csv files to dataframe and put dataframs into a list
df_list = [pd.read_csv(f"./Resources/NYC-data/{file}") for file in file_list]    
```

```python
# Combine all dfs together
df = pd.concat(df_list)
df.head()
```

```python
df.describe()
```

```python
# Drop na columns
df = df. dropna()
```

```python
# Drop duplicates
df = df.drop_duplicates()
```

```python
# Save df to csv file
df.to_csv('./Resources/nyc_citybike_data_2019.csv')
```
